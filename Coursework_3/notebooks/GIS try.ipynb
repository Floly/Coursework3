```python
result = requests.get('https://catalog.api.2gis.com/3.0/items?q=аптека&radius=2000&sort_point=37.532086,55.807215&key=ruvfds8027')
```

```python
def extract(requests_result):
    res = pd.Series(pd.Series((pd.Series(eval(requests_result.content).values())[1]).values())[0])
    Res = []
    for i in range(len(res)):
        Res.append(res[i])
    df = pd.DataFrame(Res)
    return(df, type(res))
```

```python
Aer, l = extract(result)
```
